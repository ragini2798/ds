```python
dataset = []
label = []
size = 64
```

```python
for x, image_name in enumerate(parasitized_images):
    if image_name.split(".")[1]=="png":
        image=cv2.imread(image_directory+"/parasitized/"+image_name)
        image = Image.fromarray(image,"RGB")
        image= image.resize((size,size))
        dataset.append(np.array(image))
        label.append(0)
```

```python
for y, img_name in enumerate(unparasitized_images):
    if img_name.split(".")[1]=="png":
        img=cv2.imread(image_directory+"/unparasitized/"+img_name)
        img=Image.fromarray(img,"RGB")
        img=img.resize((size,size))
        dataset.append(np.array(img))
        label.append(1)
```

```python
input_shape=(size,size,3)
inp = layers.Input(shape=input_shape)

con1 = layers.Conv2D(32, kernel_size=(3,3),activation = "relu",padding="same")(inp)

max1 = layers.MaxPooling2D(pool_size= (2,2))(con1)

norm1 = layers.BatchNormalization(axis=-1)(max1)

drop1 =layers.Dropout(rate=0.20)(norm1)
```

```python
con2 = layers.Conv2D(32, kernel_size = (3,3), activation ="relu",padding ="same")(drop1)

max2 = layers.MaxPooling2D(pool_size=(2,2))(con2)

norm2 =layers.BatchNormalization(axis=-1)(max2)

drop2= layers.Dropout(rate = 0.20)(norm2)
```

```python
flat = layers.Flatten()(drop2)
```

```python
hidden1= layers.Dense(512,activation="relu")(flat)

norm3 = layers.BatchNormalization(axis=-1)(hidden1)

drop3 = layers.Dropout(rate=0.20)(norm3)
```

```python
hidden2 = layers.Dense(256,activation="relu")(drop3)

norm4 = layers.BatchNormalization(axis=-1)(hidden2)

drop4 = layers.Dropout(rate = 0.20)(norm4)
```

```python
out = layers.Dense(2, activation ="sigmoid")(drop4)
```

```python
model = keras.Model(inputs=inp,outputs=out)
```
